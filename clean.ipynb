```python
import pandas as pd
import requests
import re

# Fetch all Pokemon species names from PokeAPI
print("Fetching all Pokemon species from PokeAPI...")
response = requests.get("https://pokeapi.co/api/v2/pokemon-species?limit=1500")
species_data = response.json()['results']

# Map API names to card-style names
SPECIAL_NAMES = {
    'nidoran-f': 'Nidoran \u2640', 'nidoran-m': 'Nidoran \u2642',
    'mr-mime': 'Mr. Mime', 'mime-jr': 'Mime Jr.', 'mr-rime': 'Mr. Rime',
    'ho-oh': 'Ho-Oh', 'porygon-z': 'Porygon-Z', 'type-null': 'Type: Null',
    'jangmo-o': 'Jangmo-o', 'hakamo-o': 'Hakamo-o', 'kommo-o': 'Kommo-o',
    'tapu-koko': 'Tapu Koko', 'tapu-lele': 'Tapu Lele',
    'tapu-bulu': 'Tapu Bulu', 'tapu-fini': 'Tapu Fini',
    'wo-chien': 'Wo-Chien', 'chien-pao': 'Chien-Pao',
    'ting-lu': 'Ting-Lu', 'chi-yu': 'Chi-Yu',
    'great-tusk': 'Great Tusk', 'scream-tail': 'Scream Tail',
    'brute-bonnet': 'Brute Bonnet', 'flutter-mane': 'Flutter Mane',
    'slither-wing': 'Slither Wing', 'sandy-shocks': 'Sandy Shocks',
    'iron-treads': 'Iron Treads', 'iron-bundle': 'Iron Bundle',
    'iron-hands': 'Iron Hands', 'iron-jugulis': 'Iron Jugulis',
    'iron-moth': 'Iron Moth', 'iron-thorns': 'Iron Thorns',
    'roaring-moon': 'Roaring Moon', 'iron-valiant': 'Iron Valiant',
    'walking-wake': 'Walking Wake', 'iron-leaves': 'Iron Leaves',
    'gouging-fire': 'Gouging Fire', 'raging-bolt': 'Raging Bolt',
    'iron-boulder': 'Iron Boulder', 'iron-crown': 'Iron Crown',
    'farfetchd': "Farfetch'd", 'sirfetchd': "Sirfetch'd",
    'flabebe': 'Flab\u00e9b\u00e9',
}

pokemon_names = set()
for entry in species_data:
    api_name = entry['name']
    if api_name in SPECIAL_NAMES:
        card_name = SPECIAL_NAMES[api_name]
    else:
        card_name = api_name.replace('-', ' ').title()
    pokemon_names.add(card_name)

# Add common card name variations
pokemon_names.update(['Nidoran', "Farfetch'd", 'Ho-Oh', 'Mime Jr', 'Unown'])

print(f"Loaded {len(pokemon_names)} Pokemon species names")
```

```python
# Build one regex matching any Pokemon name (longest first to avoid partial matches)
pokemon_pattern = re.compile(
    '|'.join(re.escape(name) for name in sorted(pokemon_names, key=len, reverse=True)),
    re.IGNORECASE,
)


def is_pokemon(species_str):
    if not species_str or pd.isna(species_str):
        return False
    return bool(pokemon_pattern.search(str(species_str)))


# Load and filter
df = pd.read_csv("pokemon_all_sorted.csv")
print(f"Total cards before cleaning: {len(df)}")

mask = df['Species'].apply(is_pokemon)
removed = df[~mask]

print(f"\nRemoving {len(removed)} non-Pokemon entries. Unique species removed:")
for species in sorted(removed['Species'].unique()):
    print(f"  \u2717 {species}")

df_clean = df[mask].reset_index(drop=True)
df_clean.to_csv("pokemon_all_cleaned.csv", index=False)
print(f"\nKept {len(df_clean)} real Pokemon cards \u2192 saved to 'pokemon_all_cleaned.csv'")
df_clean.head(10)
```

```python
df = pd.read_csv("pokemon_all_cleaned.csv")
print(f"Cards before name cleaning: {len(df)}")

# Step 1: Remove sealed products (Boxes, Tins, Collections, Decks — not actual cards)
product_pattern = re.compile(
    r'\b(Box|Tin|Collection|Deck|Premium|Chest|Bundle|Lunchbox|Mini Tin|V-UNION Special)\b',
    re.IGNORECASE,
)
is_product = df['Species'].apply(lambda s: bool(product_pattern.search(str(s))) if pd.notna(s) else False)
print(f"\nRemoving {is_product.sum()} sealed product entries:")
for sp in sorted(df.loc[is_product, 'Species'].unique()):
    print(f"  \u2717 {sp}")
df = df[~is_product].reset_index(drop=True)


# Step 2: Clean species names to core Pokemon name
def clean_species(name):
    if pd.isna(name):
        return name
    s = str(name).strip().strip('"')

    # "_____'s Mew" → "Mew", "Ash's Pikachu" → "Pikachu"
    s = re.sub(r"^[\w_]+['\u2019]s\s+", "", s)

    # "Catch Mew" → "Mew"
    s = re.sub(r'^Catch\s+', '', s)

    # For "&" Tag Team entries, clean each part individually
    if ' & ' in s:
        parts = [clean_single(p.strip()) for p in s.split(' & ')]
        return ' & '.join(parts)

    return clean_single(s)


def clean_single(s):
    # Strip regional prefixes
    s = re.sub(r'^(Alolan|Galarian|Hisuian|Paldean|Bloodmoon|Ash-)\s*', '', s)

    # Strip card variant prefixes
    s = re.sub(r'^(Dark|Light|Shining|Crystal|Ancient|Armored|Shadow Rider|Shadow|Flying|Surfing|Birthday|Black|White)\s+', '', s)

    # Strip suffixes: V, VMAX, VSTAR, GX, EX, LV.X, G, C, X, ex, V-UNION
    # Run twice to catch "G LV.X" → strip G → strip LV.X
    for _ in range(2):
        s = re.sub(r'\s+(V|VMAX|VSTAR|GX|EX|LV\.X|G|C|X|V-UNION|ex)\s*$', '', s)

    # Fix concatenated "ex" from scraping (e.g. "Alakazamex" → "Alakazam")
    if s.endswith('ex') and len(s) > 4 and s[:-2].isalpha():
        s = s[:-2]

    # "Pokemon Treasure Chest ..." or "Pokemon ..." prefix
    s = re.sub(r'^Pokemon\s+', '', s)

    return s.strip()


# Apply cleaning and track changes
df['Species_Original'] = df['Species']
df['Species'] = df['Species'].apply(clean_species)

# Show all changes
changes = df[df['Species'] != df['Species_Original']][['Species_Original', 'Species']].drop_duplicates()
print(f"\nCleaned {len(changes)} unique species names:")
for _, row in changes.sort_values('Species_Original').iterrows():
    print(f"  {row['Species_Original']} \u2192 {row['Species']}")

df = df.drop(columns=['Species_Original'])
df.to_csv("pokemon_all_cleaned.csv", index=False)
print(f"\nSaved {len(df)} cards to 'pokemon_all_cleaned.csv'")
df.head(10)
```
